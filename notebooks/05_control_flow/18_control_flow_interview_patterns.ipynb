# control_flow_interview_patterns
Interview control-flow problems usually test more than syntax. They test whether you can reason about branching, loop invariants, stopping conditions, complexity, and boundary handling under pressure.

## 1. Problem First
Why does this matter in real systems?
- Production bugs often come from the same weaknesses interviewers probe: missed cases, weak invariants, and bad exits.
- Engineers who can explain control flow clearly usually debug faster.
- Interview-style patterns map directly to scanners, validators, and state machines.

```python
numbers = [3, 7, 2, 9]
largest = numbers[0]
for number in numbers[1:]:
    if number > largest:
        largest = number
print(largest)
```

## 2. Minimal Concept
Common interview control-flow patterns:
- Running maximum or minimum
- Search until found
- Two-branch validation
- Counting with conditions
- Sliding or bounded iteration logic

## 3. Mental Model
How Python actually behaves
Most interview problems are state-update problems disguised as loops. At every step, you maintain an invariant: current best value, seen elements, valid prefix, remaining budget, or current window. If you lose the invariant, the code becomes guesswork.

```python
values = [4, 1, 8, 3]
smallest = values[0]
for value in values[1:]:
    if value < smallest:
        smallest = value
print(smallest)
```

## 4. Internal Mechanics
The mechanics are ordinary loops and branches, but the reasoning lives in the invariant. Good candidates can explain what must remain true after each iteration and why the update preserves correctness.

```python
import dis

def first_negative(values):
    for value in values:
        if value < 0:
            return value
    return None

dis.dis(first_negative)
print(first_negative([1, 2, -5, 3]))
```

## 5. Edge Cases
Where it breaks:
- Empty inputs often violate assumptions about initial state.
- Duplicate values can expose weak search or counting logic.
- Off-by-one boundaries appear in slice and range-based solutions.
- A loop may work on the happy path but fail on the first or last element.

```python
def bad_max(values):
    current = 0
    for value in values:
        if value > current:
            current = value
    return current

print(bad_max([-5, -2, -9]))
```

## 6. Performance Thinking
- Many interview control-flow problems are intentionally solvable in O(n).
- Nested loops often signal a brute-force baseline that can be improved.
- Space tradeoffs like sets or dictionaries often convert repeated scans into faster lookups.
- A correct O(n) invariant-based loop is usually better than a clever but fragile shortcut.

## 7. Real Use Case
Finding the first bad record, largest latency, or earliest failure marker in a stream are all direct analogs of interview patterns.

```python
latencies = [120, 98, 150, 110]
max_latency = latencies[0]
for latency in latencies[1:]:
    if latency > max_latency:
        max_latency = latency
print(max_latency)
```

## 8. Anti-Pattern
What beginners do wrong:
- Start coding before stating the invariant.
- Pick unsafe initial values like `0` for max problems.
- Ignore empty input handling.
- Memorize solutions without understanding the state updates.

```python
def count_even(values):
    count = 0
    for value in values:
        if value % 2 == 0:
            count += 1
    return count

print(count_even([1, 2, 3, 4, 6]))
```

## 9. Interview Signals
Questions FAANG asks:
- What invariant does your loop maintain?
- What happens on empty input?
- Can the brute-force nested-loop version be improved?
- How do you prove the final answer is correct?
