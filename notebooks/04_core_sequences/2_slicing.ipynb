# Slicing in Python

## Time: 25 mins
## Level: Beginner → Intermediate
## Prerequisite: Indexing

## 1. Problem

How do you extract parts of data efficiently?

Why is slicing powerful but sometimes expensive?

## 2. Visual Diagram (Basic Slicing)

List:

| Index | 0 | 1 | 2 | 3 | 4 | 5 |
|------|---|---|---|---|---|---|
| Value | 10 | 20 | 30 | 40 | 50 | 60 |

### Syntax

`lst[start:end:step]`

### Examples

- `lst[1:4]` → [20, 30, 40]
- `lst[:3]` → [10, 20, 30]
- `lst[::2]` → [10, 30, 50]
- `lst[::-1]` → reverse list

```python
lst = [10, 20, 30, 40, 50, 60]

print(lst[1:4])
print(lst[:3])
print(lst[::2])
print(lst[::-1])
```

## 3. Mental Model

- Start index included
- End index excluded
- Step controls jump size
- Negative step reverses direction

## 4. Internal Mechanics

```python
lst = [1, 2, 3, 4]

slice_lst = lst[:2]

print(id(lst))
print(id(slice_lst))
```

Slicing creates a NEW object.

It copies data.

## 5. Edge Cases

```python
lst = [1, 2, 3, 4, 5]

print(lst[100:])
print(lst[::-1])
print(lst[::0])  # ValueError
```

Step cannot be zero.

## 6. Performance Thinking

- Indexing → O(1)
- Slicing → O(n)

Because slicing copies elements.

## 7. Real Use Case

```python
# pagination
data = list(range(100))

page = data[0:10]
print(page)
```

Pagination and batching use slicing heavily.

## 8. Anti-Pattern

```python
# BAD
large = list(range(1000000))
copy = large[:]  # unnecessary copy
```

Large slicing wastes memory.

## 9. Interview Signals

- Why slicing is O(n)
- Difference between indexing and slicing
- Reverse list using slicing
