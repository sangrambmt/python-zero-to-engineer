# Indexing in Python



## 1. Problem

How does Python access elements instantly?

Why is `lst[0]` fast but searching is slow?

## 2. Visual Diagram (Indexing)

List:

Index:     0    1    2    3
          ┌────┬────┬────┬────┐
Values:   │ 10 │ 20 │ 30 │ 40 │
          └────┴────┴────┴────┘

Negative Index:

Index:    -4   -3   -2   -1
          ┌────┬────┬────┬────┐
Values:   │ 10 │ 20 │ 30 │ 40 │
          └────┴────┴────┴────┘

Access:
- lst[0]  → 10
- lst[2]  → 30
- lst[-1] → 40

```python
lst = [10, 20, 30, 40]

print(lst[0])
print(lst[2])
print(lst[-1])
```

## 3. Mental Model

- Index = position in memory
- Lists are ordered
- Negative index = reverse lookup

## 4. Internal Mechanics

```python
import sys

lst = [1, 2, 3]

print(sys.getsizeof(lst))
print(id(lst))
```

Lists store references in contiguous memory → O(1) access.

## 5. Edge Cases

```python
lst = [1, 2, 3]

# print(lst[10])  # IndexError

print(lst[-1])
print(lst[-2])
```

## 6. Performance Thinking

- Index → O(1)
- Search → O(n)

## 7. Real Use Case

```python
line = "ERROR 500 server_down"
parts = line.split()

status = parts[1]
print(status)
```

## 8. Anti-Pattern

```python
# BAD
lst = [1,2,3]
for i in range(len(lst)):
    print(lst[i])

# GOOD
for item in lst:
    print(item)
```
