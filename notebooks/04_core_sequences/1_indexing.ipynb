# Indexing in Python



## 1. Problem

How does Python access elements instantly in large datasets?

Why is `lst[0]` fast but searching is slow?

```python
lst = [10, 20, 30, 40]

print(lst[0])
print(lst[2])
print(lst[-1])
```

## 2. Mental Model

- Index = position in memory
- Lists are ordered
- Negative index = reverse access

`0 → first`, `-1 → last`

## 3. Internal Mechanics

```python
import sys

lst = [1, 2, 3]

print(sys.getsizeof(lst))
print(id(lst))
```

Lists store references in contiguous memory → O(1) indexing.

## 4. Edge Cases

```python
lst = [1, 2, 3]

# print(lst[10])  # IndexError

print(lst[-1])
print(lst[-2])
```

Out of range → IndexError.

## 5. Performance Thinking

- Index access → O(1)
- Search (`x in list`) → O(n)

This difference is critical at scale.

## 6. Real Use Case

```python
# log parsing
line = "ERROR 500 server_down"

parts = line.split()

status = parts[1]
print(status)
```

Indexing after split is common in real systems.

## 7. Anti-Pattern

```python
# BAD
lst = [1,2,3]
for i in range(len(lst)):
    print(lst[i])

# GOOD
for item in lst:
    print(item)
```

Avoid index-based loops unless needed.
