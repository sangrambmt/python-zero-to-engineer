# Indexing in Python



## 1. Problem

How does Python access elements instantly?

Why is `lst[0]` fast but searching is slow?

## 2. Visual Diagram (Indexing)

### Positive Indexing

| Index | 0 | 1 | 2 | 3 |
|------|---|---|---|---|
| Value | 10 | 20 | 30 | 40 |

### Negative Indexing

| Index | -4 | -3 | -2 | -1 |
|------|----|----|----|----|
| Value | 10 | 20 | 30 | 40 |

### Access
- lst[0]  → 10
- lst[2]  → 30
- lst[-1] → 40

```python
lst = [10, 20, 30, 40]

print(lst[0])
print(lst[2])
print(lst[-1])
```

## 3. Mental Model

- Index = position in memory
- Lists are ordered
- Negative index = reverse lookup
- Same structure, different direction

## 4. Internal Mechanics

```python
import sys

lst = [1, 2, 3]

print(sys.getsizeof(lst))
print(id(lst))
```

Lists store references in contiguous memory → O(1) access.

## 5. Edge Cases

```python
lst = [1, 2, 3]

# print(lst[10])  # IndexError

print(lst[-1])
print(lst[-2])
```

Out of range → IndexError.

## 6. Performance Thinking

- Index → O(1)
- Search → O(n)

Critical difference at scale.

## 7. Real Use Case

```python
line = "ERROR 500 server_down"
parts = line.split()

status = parts[1]
print(status)
```

Indexing after split is common in real systems.

## 8. Anti-Pattern

```python
# BAD
lst = [1,2,3]
for i in range(len(lst)):
    print(lst[i])

# GOOD
for item in lst:
    print(item)
```

Avoid index-based loops unless necessary.
