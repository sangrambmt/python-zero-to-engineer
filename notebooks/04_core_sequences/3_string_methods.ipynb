# String Methods in Python

## 1. Problem

Why do real-world systems spend huge time processing strings?

Logs, APIs, CSVs, user input and files are mostly text.

## 2. Visual Diagram

| Method | Purpose |
|--------|---------|
| lower() | lowercase |
| upper() | uppercase |
| strip() | remove spaces |
| split() | split text |
| replace() | replace values |
| join() | combine strings |
| find() | search substring |
| startswith() | prefix check |
| endswith() | suffix check |

```python
text = "  Python Engineering  "

print(text.lower())
print(text.upper())
print(text.strip())
```

## 3. Mental Model

- Strings are immutable
- Methods create NEW strings
- Original string remains unchanged

## 4. Internal Mechanics

```python
s = "hello"

new_s = s.upper()

print(id(s))
print(id(new_s))
```

New object created because strings are immutable.

## 5. Common Methods

```python
text = "python,data,engineering"

print(text.split(','))
print(text.replace('python', 'AI'))
print(text.find('data'))
```

split() is heavily used in parsing.

```python
words = ['Python', 'is', 'powerful']

sentence = ' '.join(words)
print(sentence)
```

join() is more efficient than repeated string concatenation.

## 6. Edge Cases

```python
text = ""

print(text.strip())
print(text.split(','))
print(text.find('a'))
```

find() returns -1 when substring not found.

## 7. Performance Thinking

- String concatenation in loops is expensive
- join() is optimized
- String methods usually create copies

## 8. Real Use Case

```python
# log parsing
log = "ERROR: Server Down"

level = log.split(':')[0]
message = log.split(':')[1].strip()

print(level)
print(message)
```

Text parsing is everywhere in engineering systems.

## 9. Anti-Pattern

```python
# BAD
result = ""

for i in ['a', 'b', 'c']:
    result += i

print(result)
```

Repeated concatenation creates many temporary strings.

```python
# GOOD
chars = ['a', 'b', 'c']
print(''.join(chars))
```
