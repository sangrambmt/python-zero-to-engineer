# Iteration Basics in Python



## 1. Problem

How do systems process millions of records one-by-one efficiently?

Iteration is the foundation of data processing.

## 2. Visual Diagram

| Position | Value |
|----------|-------|
| 0 | 10 |
| 1 | 20 |
| 2 | 30 |
| 3 | 40 |

Iteration Flow:

10 → 20 → 30 → 40

```python
numbers = [10, 20, 30, 40]

for num in numbers:
    print(num)
```

## 3. Mental Model

- Iteration processes elements sequentially
- Python uses iterators internally
- `for` loop hides iterator complexity

## 4. Internal Mechanics

```python
numbers = [1, 2, 3]

iterator = iter(numbers)

print(next(iterator))
print(next(iterator))
print(next(iterator))
```

`for` loop internally calls iter() and next().

## 5. Iterating Different Structures

```python
text = "Python"

for ch in text:
    print(ch)
```

```python
data = {'a': 1, 'b': 2}

for key, value in data.items():
    print(key, value)
```

Most Python collections are iterable.

## 6. Edge Cases

```python
iterator = iter([1])

print(next(iterator))

# print(next(iterator))  # StopIteration
```

Iterator exhaustion raises StopIteration.

## 7. Performance Thinking

- Iteration is O(n)
- Generators reduce memory usage
- Avoid unnecessary nested loops

## 8. Real Use Case

```python
# log processing
logs = [
    'ERROR server down',
    'INFO user login',
    'WARNING memory high'
]

for log in logs:
    if 'ERROR' in log:
        print(log)
```

Real systems iterate over streams, logs and records.

## 9. Anti-Pattern

```python
# BAD
lst = [1,2,3]

for i in range(len(lst)):
    print(lst[i])
```

Avoid index iteration unless index is needed.

```python
# GOOD
for item in lst:
    print(item)
```
