# Operator Precedence

## 1. Problem

Why does Python give unexpected results when multiple operators are used together?

```python
print(2 + 3 * 4)      # 14
print((2 + 3) * 4)    # 20
```

## 2. Mental Model

Python follows precedence rules (similar to math):

1. Parentheses ()
2. Exponentiation **
3. Multiplication, Division *, /, //, %
4. Addition, Subtraction +, -
5. Comparisons
6. Logical operators

## 3. Internal Mechanics

```python
import dis

def expr():
    return 2 + 3 * 4

dis.dis(expr)
```

Multiplication happens before addition.

## 4. Edge Cases

```python
print(True + True * 2)   # 3
print(10 > 5 and 2 < 1) # False
print(10 > (5 and 2) < 1) # tricky
```

Boolean values behave like integers.

Logical precedence can confuse expressions.

## 5. Performance Thinking

- Precedence does not change complexity
- But wrong grouping leads to incorrect logic

## 6. Real Use Case

```python
price = 100
discount = 0.2

final_price = price * (1 - discount)
print(final_price)
```

Parentheses ensure correct business logic.

## 7. Anti-Pattern

```python
# BAD
result = 2 + 3 * 4 + 5 * 6  # hard to read

# GOOD
result = (2 + (3 * 4)) + (5 * 6)
```
