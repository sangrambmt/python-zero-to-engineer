# Comparison Operators

## 1. Problem

Why does Python sometimes compare values that look the same but behave differently?

```python
a = 10
b = 20

print(a == b)
print(a != b)
print(a > b)
print(a < b)
print(a >= b)
print(a <= b)
```

## 2. Mental Model

- Comparison returns Boolean (True/False)
- Used in decision making
- Works across types (sometimes unexpectedly)

## 3. Internal Mechanics

```python
import dis

def compare(a, b):
    return a > b

dis.dis(compare)
```

## 4. Edge Cases

```python
print(10 == "10")   # False
print(10 == 10.0)   # True

# print(10 > "5")   # TypeError in Python 3
```

## 5. Performance Thinking

- Comparisons are O(1)
- But string comparisons depend on length

## 6. Real Use Case

```python
score = 85

if score >= 90:
    print("Grade A")
elif score >= 75:
    print("Grade B")
else:
    print("Grade C")
```

## 7. Anti-Pattern

```python
# BAD
if x == True:
    pass

# GOOD
if x:
    pass
```
