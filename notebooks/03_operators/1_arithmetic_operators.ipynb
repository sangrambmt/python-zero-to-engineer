# Arithmetic Operators

## 1. Problem

Why does integer division behave differently from float division?

```python
a = 10
b = 3

print(a + b)
print(a - b)
print(a * b)
print(a / b)   # float division
print(a // b)  # integer division
print(a % b)   # remainder
print(a ** b)  # power
```

## 2. Mental Model

- `/` always returns float
- `//` floors the result
- `%` gives remainder
- `**` exponentiation

## 3. Internal Mechanics

```python
import dis

def add(a, b):
    return a + b

dis.dis(add)
```

## 4. Edge Cases

```python
print(-10 // 3)   # -4 (floor behavior)
print(10 % 3)     # 1

# division by zero
# print(10 / 0)  # ZeroDivisionError
```

## 5. Performance Thinking

- All arithmetic ops are O(1)
- Division is slower than addition/multiplication

## 6. Real Use Case

```python
# pagination logic
total_items = 105
page_size = 10

pages = total_items // page_size + (total_items % page_size > 0)
print(pages)
```

## 7. Anti-Pattern

```python
# BAD
a = 10
b = 3

print(int(a / b))  # incorrect rounding behavior
```
