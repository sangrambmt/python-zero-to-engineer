# Membership Operators

## 1. Problem

Why is checking membership in a list slow but fast in a set?

```python
lst = [1, 2, 3, 4]

print(2 in lst)
print(5 not in lst)
```

## 2. Mental Model

- `in` checks presence
- `not in` checks absence
- Works on sequences and collections

## 3. Internal Mechanics

```python
lst = list(range(1000))

for i in lst:
    if i == 999:
        print("Found")
```

List search is linear scan (O(n)).

```python
s = set(range(1000))

print(999 in s)
```

Set uses hashing (O(1)).

## 4. Edge Cases

```python
print('a' in 'apple')
print('z' in 'apple')

d = {'a': 1, 'b': 2}
print('a' in d)  # checks keys only
```

Dictionary membership checks keys, not values.

## 5. Performance Thinking

- list → O(n)
- set → O(1)
- dict → O(1) for keys

## 6. Real Use Case

```python
allowed_roles = {'admin', 'user'}

role = 'admin'

if role in allowed_roles:
    print("Access granted")
```

## 7. Anti-Pattern

```python
# BAD
lst = [1,2,3,4,5]

if 5 in lst: pass  # inefficient for large data

# BETTER
s = set(lst)
if 5 in s: pass
```

## 8. Interview Signals

- Why set is faster than list
- Membership complexity
- Dictionary key lookup behavior
