# Identity Operators

## 1. Problem

Why do two equal values sometimes not share the same identity?

```python
a = [1, 2]
b = a
c = [1, 2]

print(a is b)
print(a is c)
print(a == c)
```

## 2. Mental Model

- `is` checks memory identity
- `==` checks value equality
- Same value ≠ same object

## 3. Internal Mechanics

```python
a = [1, 2]
b = a

print(id(a))
print(id(b))
```

Same memory reference.

```python
c = [1, 2]
print(id(c))
```

Different object, same value.

## 4. Edge Cases

```python
x = 256
y = 256

print(x is y)  # often True (cached)

x = 1000
y = 1000

print(x is y)  # often False
```

Small integers may be cached by Python.

## 5. Performance Thinking

- `is` is faster than `==`
- But only use for identity checks

## 6. Real Use Case

```python
value = None

if value is None:
    print("No value")
```

`is None` is the correct pattern.

## 7. Anti-Pattern

```python
# BAD
x = 1000
y = 1000

print(x is y)  # unreliable

# GOOD
print(x == y)
```

## 8. Interview Signals

- Difference between `is` and `==`
- Why integer caching exists
- When to use `is None`
