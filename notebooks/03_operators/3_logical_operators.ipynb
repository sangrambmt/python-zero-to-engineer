# Assignment Operators

## 1. Problem

Why does `+=` behave differently for lists and integers?

```python
x = 10
x += 5
print(x)

x -= 3
print(x)

x *= 2
print(x)

x /= 4
print(x)
```

## 2. Mental Model

- `+=`, `-=`, `*=`, `/=` modify variable
- Behavior depends on mutability
- Immutable → new object
- Mutable → modify in-place

## 3. Internal Mechanics

```python
a = [1, 2]
b = a

a += [3]

print(a)
print(b)  # both changed
```

`+=` modifies list in-place.

```python
x = 10
y = x

x += 5

print(x)
print(y)  # unchanged
```

Immutable types create new object.

## 4. Edge Cases

```python
a = (1, 2)
b = a

a += (3,)

print(a)
print(b)  # unchanged
```

Tuple is immutable → new object created.

## 5. Performance Thinking

- In-place operations save memory
- Immutable operations create new objects

## 6. Real Use Case

```python
total = 0

for i in range(5):
    total += i

print(total)
```

## 7. Anti-Pattern

```python
# BAD
a = [1, 2]
b = a
a = a + [3]  # new object, breaks reference

print(a)
print(b)
```

## 8. Interview Signals

- Difference between `+=` and `+`
- Mutable vs immutable behavior
- Memory implications
