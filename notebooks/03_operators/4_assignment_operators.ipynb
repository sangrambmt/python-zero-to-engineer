# Assignment Operators

Variable update and in-place modification patterns.

## Problem

Why does `+=` behave differently for lists and integers?

Understanding assignment behavior is critical for debugging shared-state bugs.

## Visual Diagram

| Operator | Meaning |
|----------|---------|
| = | assign |
| += | add and assign |
| -= | subtract and assign |
| *= | multiply and assign |
| /= | divide and assign |
| //= | floor divide and assign |
| %= | modulo and assign |
| **= | power and assign |

```python
x = 10

x += 5
print(x)

x -= 3
print(x)

x *= 2
print(x)

x /= 4
print(x)
```

## Mental Model

- Assignment updates references
- Immutable objects create new memory
- Mutable objects may update in-place

## Internal Mechanics

```python
a = 10

print(id(a))

a += 5

print(id(a))
```

Integers are immutable.

New object gets created.

```python
lst = [1, 2]

print(id(lst))

lst += [3]

print(id(lst))
```

Lists are mutable.

Modification happens in-place.

## Edge Cases

```python
a = [1, 2]
b = a

a += [3]

print(a)
print(b)
```

Both references change because list is shared.

```python
x = (1, 2)
y = x

x += (3,)

print(x)
print(y)
```

Tuple creates new object because it is immutable.

## Performance Thinking

- In-place updates reduce allocations
- Immutable updates create new objects
- Large object copies are expensive

## Real Use Case

```python
# accumulator pattern
total = 0

for num in [10, 20, 30]:
    total += num

print(total)
```

Assignment operators are heavily used in aggregation systems.

## Anti-Pattern

```python
# BAD
result = result + value

# BETTER
result += value
```

Use compound operators for readability and efficiency.
