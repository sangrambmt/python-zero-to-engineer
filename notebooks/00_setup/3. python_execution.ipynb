# Python Execution Model

This notebook explains how Python runs code.

Not syntax.  
Not basics.  
Execution.

Goal:
- Understand interpreter
- Understand bytecode
- Understand memory behavior

This is where most people stay shallow.

## 1. Python is an Interpreted Language

Python does NOT run code directly.

Steps:
1. Source code (.py)
2. Compiled to bytecode (.pyc)
3. Executed by Python Virtual Machine (PVM)

We will prove this.

```python
# Simple code
x = 10
y = 20
print(x + y)
```

## 2. Inspect Bytecode

Python converts your code into bytecode.

We can inspect it using `dis` module.

```python
import dis

def add(a, b):
    return a + b

dis.dis(add)
```

Observe:
- LOAD_FAST
- BINARY_OP
- RETURN_VALUE

This is what Python actually executes.

Not your code.

## 3. Compilation Step (Behind the Scenes)

Python compiles functions into bytecode objects.

```python
def multiply(a, b):
    return a * b

print(multiply.__code__)
```

## 4. Python Memory Model

Everything in Python is an object.

Even integers.

```python
x = 10
y = 10

print(id(x), id(y))
```

Same memory reference.

Python reuses small integers.

```python
a = 1000
b = 1000

print(id(a), id(b))
```

Different memory references.

This behavior matters in performance and bugs.

## 5. Reference vs Value

Variables are references, not containers.

```python
a = [1, 2, 3]
b = a

b.append(4)

print(a)  # changed
```

Both `a` and `b` point to the same object.

This is a common source of bugs.

## 6. Immutable vs Mutable Impact

```python
x = 10
y = x

y += 1

print(x)  # unchanged
print(y)
```

Integers are immutable.

New object created.

## 7. Execution Flow

Python executes top to bottom.

But function bodies execute only when called.

```python
print("Start")

def test():
    print("Inside function")

print("End")

test()
```

## 8. Performance Insight

Python is slower than C because:
- Dynamic typing
- Interpreted execution
- Object overhead

But:
- Faster to write
- Easier to debug

## 9. Real Use Case

Understanding execution helps in:
