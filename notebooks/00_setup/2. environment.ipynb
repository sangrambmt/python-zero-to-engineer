## 14. Environment Validation (Reality Check)

Before running any project, verify:
- Python version
- Installed packages
- Active environment

This prevents “works on my machine” issues.

```python
import sys
import pkg_resources

print("Python Version:", sys.version)
print("Executable Path:", sys.executable)

print("\nInstalled Packages:")
for pkg in sorted([p.project_name + "==" + p.version for p in pkg_resources.working_set]):
    print(pkg)
```
