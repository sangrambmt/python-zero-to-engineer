# list_comprehensions

```python
# basic syntax
values = [1, 2, 3, 4, 5]
squares = [x * x for x in values]
print(squares)
```

```python
# vs equivalent for loop
values = [1, 2, 3, 4, 5]
squares = []
for x in values:
    squares.append(x * x)
print(squares)
```

```python
# with condition (filter)
latencies = [90, 120, 150, 80, 200]
slow = [x for x in latencies if x > 100]
print(slow)
```

```python
# transform and filter together
latencies = [90, 120, 150, 80, 200]
slow_ms = [f"{x}ms" for x in latencies if x > 100]
print(slow_ms)
```

```python
# if-else expression (ternary)
latencies = [90, 120, 150, 80, 200]
labels = ["slow" if x > 100 else "fast" for x in latencies]
print(labels)
```

```python
# string operations
services = ["  api  ", "  worker  ", "  scheduler  "]
cleaned = [s.strip() for s in services]
print(cleaned)
```

```python
# string transform
services = ["api", "worker", "scheduler"]
upper = [s.upper() for s in services]
print(upper)
```

```python
# from range
evens = [x for x in range(10) if x % 2 == 0]
print(evens)
```

```python
# from string
letters = [ch for ch in "pipeline" if ch not in "aeiou"]
print(letters)
```

```python
# from dict items
latencies = {"api": 120, "worker": 95, "scheduler": 300}
slow_services = [k for k, v in latencies.items() if v > 100]
print(slow_services)
```

```python
# nested list comprehension (flatten)
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flat = [val for row in matrix for val in row]
print(flat)
```

```python
# nested list comprehension (build matrix)
matrix = [[row * col for col in range(1, 4)] for row in range(1, 4)]
for row in matrix:
    print(row)
```

```python
# multiple for clauses
regions = ["us", "eu"]
services = ["api", "worker"]
deployments = [f"{r}-{s}" for r in regions for s in services]
print(deployments)
```

```python
# with enumerate
services = ["api", "worker", "scheduler"]
indexed = [f"{i}: {s}" for i, s in enumerate(services, start=1)]
print(indexed)
```

```python
# with zip
services = ["api", "worker", "scheduler"]
statuses = ["running", "stopped", "running"]
pairs = [f"{s} is {st}" for s, st in zip(services, statuses)]
print(pairs)
```

```python
# function call in expression
def tag(service):
    return f"[{service}]"

services = ["api", "worker", "scheduler"]
tagged = [tag(s) for s in services]
print(tagged)
```

```python
# remove None values
values = [1, None, 2, None, 3]
clean = [x for x in values if x is not None]
print(clean)
```

```python
# deduplicate while preserving order
values = [3, 1, 4, 1, 5, 3, 2]
seen = set()
unique = [x for x in values if not (x in seen or seen.add(x))]
print(unique)
```

```python
# list comprehension vs map()
values = [1, 2, 3, 4]
via_map = list(map(lambda x: x * 2, values))
via_comp = [x * 2 for x in values]
print(via_map)
print(via_comp)
```

```python
# list comprehension vs filter()
values = [1, 2, 3, 4, 5, 6]
via_filter = list(filter(lambda x: x % 2 == 0, values))
via_comp = [x for x in values if x % 2 == 0]
print(via_filter)
print(via_comp)
```

```python
# generator expression (lazy, memory efficient)
latencies = [90, 120, 150, 80, 200]
total = sum(x for x in latencies if x > 100)
print(total)
```

```python
# set comprehension
events = ["ERROR", "INFO", "ERROR", "WARN", "ERROR"]
unique_events = {e for e in events}
print(unique_events)
```

```python
# dict comprehension
services = ["api", "worker", "scheduler"]
status = {s: "running" for s in services}
print(status)
```

```python
# walrus operator := (Python 3.8+)
latencies = [90, 120, 150, 80, 200]
results = [trimmed for x in latencies if (trimmed := x - 10) > 100]
print(results)
```

```python
# readability: when to use a for loop instead
# complex logic is clearer as a loop
results = []
for service in ["api", "worker", "scheduler"]:
    if service.startswith("a"):
        label = service.upper()
        results.append(f"active: {label}")
print(results)
```

```python
# interview problem: matrix transpose
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
transposed = [[row[i] for row in matrix] for i in range(len(matrix[0]))]
for row in transposed:
    print(row)
```
