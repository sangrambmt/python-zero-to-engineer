# list_comprehensions
List comprehensions are compact transformation and filtering tools. They can turn noisy loops into clear data-shaping expressions, but they become a readability trap when too much business logic gets compressed into one line.

## 1. Problem First
Why does this matter in real systems?
- Pipelines often transform records into normalized forms.
- Filtering and projection are extremely common in data workflows.
- A clear comprehension can reduce boilerplate; a dense one can hide bugs.

```python
latencies = [90, 120, 150]
slow = [latency for latency in latencies if latency > 100]
print(slow)
```

## 2. Minimal Concept
Core syntax:
- `[expression for item in iterable]`
- `[expression for item in iterable if condition]`
- They build a new list.

## 3. Mental Model
How Python actually behaves
A list comprehension is a compact loop plus optional filter that eagerly builds a new list in memory. It is best when the transformation is simple enough that the whole idea still fits in one readable expression.

```python
names = [" api ", "worker", " scheduler "]
normalized = [name.strip().upper() for name in names]
print(normalized)
```

## 4. Internal Mechanics
Comprehensions are compiled into their own internal scope in modern Python. They still do the same essential work as loops: iterate, filter, compute, and append into a result list.

```python
import dis

def double(values):
    return [value * 2 for value in values]

dis.dis(double)
print(double([1, 2, 3]))
```

## 5. Edge Cases
Where it breaks:
- Nested comprehensions get unreadable quickly.
- Side effects inside comprehensions are a code smell.
- Eager list construction can waste memory for large data.
- A normal loop is often clearer when multiple steps or names are needed.

```python
matrix = [[1, 2], [3, 4]]
flattened = [value for row in matrix for value in row]
print(flattened)

verbose = []
for row in matrix:
    for value in row:
        verbose.append(value)
print(verbose)
```

## 6. Performance Thinking
- Comprehensions are usually as fast as or faster than equivalent append loops.
- They are still O(n) in the number of processed items.
- Memory cost is O(n) because the full result list is built eagerly.
- If you only need streaming behavior, a generator expression may be better.

## 7. Real Use Case
A log processing step may normalize and filter messages before sending them downstream.

```python
records = [
    {"level": "INFO", "message": "boot"},
    {"level": "ERROR", "message": "db timeout"},
    {"level": "ERROR", "message": "disk full"}
]
error_messages = [record["message"].upper() for record in records if record["level"] == "ERROR"]
print(error_messages)
```

## 8. Anti-Pattern
What beginners do wrong:
- Force complex branching into one comprehension.
- Use comprehensions for side effects instead of building values.
- Choose compactness over readability.

```python
values = [1, 2, 3]
[print(value) for value in values]
print("side-effect comprehensions are usually not a good idea")
```
