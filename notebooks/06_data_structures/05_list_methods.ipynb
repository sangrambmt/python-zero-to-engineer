# list_methods
One method or topic per code cell.

```python
# append()
jobs = ["extract", "transform"]
jobs.append("load")
print(jobs)
```

```python
# extend()
jobs = ["extract", "transform"]
jobs.extend(["load", "notify"])
print(jobs)
```

```python
# insert()
steps = ["extract", "load"]
steps.insert(1, "transform")
print(steps)
```

```python
# remove()
levels = ["INFO", "DEBUG", "ERROR"]
levels.remove("DEBUG")
print(levels)
```

```python
# pop()
queue = ["job1", "job2", "job3"]
removed = queue.pop()
print(removed)
print(queue)
```

```python
# clear()
buffer = [1, 2, 3]
buffer.clear()
print(buffer)
```

```python
# index()
services = ["api", "worker", "scheduler"]
print(services.index("worker"))
```

```python
# count()
events = ["ERROR", "INFO", "ERROR", "WARN", "ERROR"]
print(events.count("ERROR"))
```

```python
# sort()
latencies = [120, 95, 300, 110]
latencies.sort()
print(latencies)
```

```python
# reverse()
checkpoints = [1, 2, 3, 4]
checkpoints.reverse()
print(checkpoints)
```

```python
# copy()
original = [1, 2, 3]
copied = original.copy()
copied.append(4)
print(original)
print(copied)
```

```python
# slicing
records = [10, 20, 30, 40, 50]
print(records[1:4])
print(records[:3])
print(records[::2])
```

```python
# list comprehension
latencies = [90, 120, 150, 80]
slow = [value for value in latencies if value > 100]
print(slow)
```

```python
# nested lists
matrix = [[1, 2], [3, 4], [5, 6]]
print(matrix[1][0])
print(matrix)
```

```python
# shallow copy vs deep copy
import copy
original = [[1], [2]]
shallow = original.copy()
deep = copy.deepcopy(original)
original[0].append(99)
print(original)
print(shallow)
print(deep)
```

```python
# membership operators
allowed = ["read", "write", "deploy"]
print("deploy" in allowed)
print("admin" not in allowed)
```

```python
# concatenation
batch_a = [1, 2, 3]
batch_b = [4, 5]
combined = batch_a + batch_b
print(combined)
```

```python
# repetition
zeros = [0] * 5
print(zeros)
```

```python
# unpacking
values = [10, 20, 30, 40]
first, second, *rest = values
print(first, second, rest)
```

```python
# max()
latencies = [120, 95, 300, 110]
print(max(latencies))
```

```python
# min()
latencies = [120, 95, 300, 110]
print(min(latencies))
```

```python
# sum()
latencies = [120, 95, 300, 110]
print(sum(latencies))
```

```python
# len()
jobs = ["extract", "transform", "load"]
print(len(jobs))
```

```python
# any()
health_checks = [False, False, True]
print(any(health_checks))
```

```python
# all()
health_checks = [True, True, True]
print(all(health_checks))
```

```python
# enumerate()
lines = ["INFO boot", "WARN disk", "ERROR db"]
for line_number, line in enumerate(lines, start=1):
    print(line_number, line)
```

```python
# zip()
services = ["api", "worker"]
regions = ["us", "eu"]
print(list(zip(services, regions)))
```

```python
# map()
values = [1, 2, 3, 4]
doubled = list(map(lambda x: x * 2, values))
print(doubled)
```

```python
# filter()
values = [1, 2, 3, 4, 5, 6]
even_values = list(filter(lambda x: x % 2 == 0, values))
print(even_values)
```

```python
# sorted()
values = [3, 1, 4, 1, 5]
ordered = sorted(values)
print(ordered)
print(values)
```

```python
# reversed()
values = [3, 1, 4, 1, 5]
print(list(reversed(values)))
print(values)
```

```python
# stack implementation
stack = []
stack.append("open")
stack.append("write")
print(stack.pop())
print(stack)
```

```python
# queue implementation
queue = ["job1", "job2", "job3"]
next_job = queue.pop(0)
print(next_job)
print(queue)
```

```python
# matrix traversal
matrix = [[1, 2], [3, 4], [5, 6]]
for row in matrix:
    for value in row:
        print(value)
```

```python
# flatten list
matrix = [[1, 2], [3, 4], [5, 6]]
flat = [value for row in matrix for value in row]
print(flat)
```

```python
# remove duplicates
values = [1, 2, 2, 3, 1, 4]
unique = list(dict.fromkeys(values))
print(unique)
```

```python
# frequency counter
events = ["ERROR", "INFO", "ERROR", "WARN", "ERROR"]
freq = {}
for event in events:
    freq[event] = freq.get(event, 0) + 1
print(freq)
```

```python
# list interview problems
def two_sum(nums, target):
    seen = {}
    for index, value in enumerate(nums):
        needed = target - value
        if needed in seen:
            return seen[needed], index
        seen[value] = index
    return None

print(two_sum([2, 7, 11, 15], 9))
```
